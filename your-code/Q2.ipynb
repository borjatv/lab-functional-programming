Now we want to enhance the `get_bow_from_docs` function so that it will work with HTML webpages. In HTML, there are a lot of messy codes such as HTML tags, Javascripts, [unicodes](https://www.w3schools.com/charsets/ref_utf_misc_symbols.asp) that will mess up your bag of words. We need to clean up those junk before generating BoW.

Next, what you will do is to define several new functions each of which is specialized to clean up the HTML codes in one aspect. For instance, you can have a `strip_html_tags` function to remove all HTML tags, a `remove_punctuation` function to remove all punctuation, a `to_lower_case` function to convert string to lowercase, and a `remove_unicode` function to remove all unicodes.

Then in your `get_bow_from_doc` function, you will call each of those functions you created to clean up the HTML before you generate the corpus.

Note: Please use Python string operations and regular expression only in this lab. Do not use extra libraries such as `beautifulsoup` because otherwise you loose the purpose of practicing.

```python
# Define your string handling functions below
# Minimal 3 functions

import re

#1 remove html tags

def strip_html_tags(txt):
    htmlout = re.compile('<[^>]+>')
    return htmlout.sub(' ', txt)

def remove_punctuation(txt):
    char = ['.',':',';','(',')','^',',','{','}','-','?','"','!','|','=',"'",'*','#','[',']','\t']
    for a in char:
        newtxt = txt.replace(a, ' ')
        txt = newtxt
    return txt

def remove_unicode(txt):
    unicodeout = re.compile('&#\d*')
    return unicodeout.sub(' ', txt)

def to_lower_case(txt):
    return txt.lower()
```

Next, paste your previously written `get_bow_from_docs` function below. Call your functions above at the appropriate place.

```python
def get_bow_from_docs (docs, stop_words=[]):
    # In the function, first define the variables you will use such as `corpus`, `bag_of_words`, and `term_freq`.
    corpus = []
    bag_of_words = []
    term_freq = []
    temp = []
    
    # write your codes here
    
    for x in docs:
        with open(x, 'r') as f:
            corpus.append(f.read())
    
    for a in corpus:
        b = strip_html_tags(a)
        c = remove_punctuation(b)
        d = remove_unicode(c)
        e = to_lower_case(d)
        temp.append(e)
        #corpus.append(to_lower_case(remove_unicode(remove_punctuation(strip_html_tags(a)))))
    print(len(temp))
    
    
    for x in temp:
        a = x.split()
        for y in a:
            if y not in bag_of_words and y not in stop_words:
                bag_of_words.append(y)
   
    for x in temp:
        a = x.split()
        eachlist = [a.count(b) for b in bag_of_words]
        term_freq.append(eachlist) 
   
    return {
        "bag_of_words": bag_of_words,
        "term_freq": term_freq
    }
```

```python
# Cadenas de más de 20 caracteres
def longchains(txt):
    htmlout = re.compile('\S{20,}\s')
    return htmlout.sub(' ', txt)
```

```python
# Cadenas en las cuales aparezca un número (ya que no añadirán valor desde el punto de vista de las palabras)
# La ejecutamos antes que las primeras
def numbers(txt):
    htmlout = re.compile('\S*\d+\S*')
    return htmlout.sub(' ', txt)
```

```python
# Cadenas en las cuales aparezcan símbolos. Esta la ponemos al principio para que se ejecute antes que la función
# que elimina los signos de puntuación: & /  = $
def symbols(txt):
    htmlout = re.compile('\S*\&+\S*|\S*\/+\S*|\S*\=+\S*|\S*\$+\S*')
    return htmlout.sub(' ', txt)
```

```python
# Cadenas que contengan el símbolo \, la ponemos al final
def removeslash(txt):
    htmlout = re.compile('\S*\\+\S*')
    return htmlout.sub(' ', txt)
```

```python
def get_bow_from_docs_improved (docs, stop_words=[]):
    # In the function, first define the variables you will use such as `corpus`, `bag_of_words`, and `term_freq`.
    corpus = []
    bag_of_words = []
    term_freq = []
    temp = []
    
    # write your codes here
    
    for x in docs:
        with open(x, 'r') as f:
            corpus.append(f.read())
    
    for r in corpus:
        t = strip_html_tags(r)
        u = longchains(t)
        a = numbers(u)
        b = symbols(a)
        c = remove_punctuation(b)
        d = remove_unicode(c)
        e = to_lower_case(d)
        f = longchains(e)
        g = numbers(f)
        h = removeslash(g)
        temp.append(h)
        #corpus.append(to_lower_case(remove_unicode(remove_punctuation(strip_html_tags(a)))))
    print(len(temp))
    
    
    for x in temp:
        a = x.split()
        for y in a:
            if y not in bag_of_words and y not in stop_words:
                bag_of_words.append(y)
   
    for x in temp:
        a = x.split()
        eachlist = [a.count(b) for b in bag_of_words]
        term_freq.append(eachlist) 
   
    return {
        "bag_of_words": bag_of_words,
        "term_freq": term_freq
    }
```
